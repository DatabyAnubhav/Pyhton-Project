```python
''' Task 2 
The cost column in the dataframe “Project” has some missing values. Your task is to compute these missing values. 
Replace the missing values by "running average". You should use the “For” loop for this task. '''

df_project = pd.read_csv('project.csv') # Read the 'project.csv' file into a DataFrame named df_project

# Initialize running_sum to store the cumulative sum of non-missing Cost values
running_sum = 0.0

# Initialize count to store the number of non-missing Cost values encountered so far
count = 0

# Loop through each row index in df_project
for i in range(len(df_project)):

    # Check if the 'Cost' value at row i is NOT NaN (i.e., it's a valid number)
    if pd.notna(df_project.loc[i,'Cost']): 
        
        # Add the current 'Cost' value to running_sum
        running_sum += df_project.loc[i, 'Cost'] 

        # Increment the count of non-missing values
        count += 1

    else:
        df_project.loc[i,'Cost'] = round(running_sum/count,2)  # If the 'Cost' value is NaN, replace it with the running mean (rounded to 2 decimal places)

df_project.to_csv('project_fill.csv', index= False) # Save the updated df_project DataFrame (with filled 'Cost' values) to 'project_fill.csv' without index column

project_fill = pd.read_csv('project_fill.csv') # Read the updated CSV file 'project_fill.csv' into a new DataFrame named project_fill

project_fill   # Display the contents of the project_fill DataFrame

```

```python
''' Task 3 
 Split the name column in the Employee dataframe into two new columns “First Name”, and “Last Name” and remove the older “name” column. '''

df_employee[['First Name', 'Last Name']] = df_employee['Name'].str.split(' ', n=1, expand=True)  # expand=True returns the result as two new columns

# Drop the original 'Name' column (axis=1 specifies column), inplace=True applies the change directly
df_employee.drop('Name', axis=1, inplace=True)

# Save the modified DataFrame (with split names) to 'employee_split.csv'
df_employee.to_csv('employee_split.csv', index=False)

df_employee
```

```python
''' Task4
Join all three dataframes in one single dataframe. Name it “Final” '''

# Merge project_fill and df_employee DataFrames on the common column 'ID'
# how='left' → Keep all rows from project_fill (even if there’s no matching employee info)
merged = pd.merge(project_fill, df_employee, on='ID', how='left')

# Merge the resulting 'merged' DataFrame with df_seniority_level_data on 'ID'
df_final = pd.merge(merged, df_seniority_level_data, on='ID')

# Save the final merged DataFrame into 'final.csv'
df_final.to_csv('final.csv')
```

```python
# Task 5 
'''Add a new bonus column in the Final dataframe. Give a 5% bonus concerning project cost only to employees who have finished the projects. '''

# Read the 'final.csv' file into a DataFrame named df_final
pd.read_csv('final.csv')

# Create a new column 'Bonus'
# For each row, apply this logic:
# → If 'Status' is 'Finished', Bonus = 5% of 'Cost'
# → Otherwise, Bonus = 0
# axis=1 means apply the lambda function **row-wise*

df_final['Bonus'] = df_final.apply(lambda x: x['Cost']*0.05 if x['Status'] == 'Finished' else 0, axis = 1) 

df_final.to_csv('final.csv')
```

```python
''' Task 6 
Demote the designation level by 1, whose projects have status “fail”. Delete the employees record whose designation level is above 4. '''

# Create a Boolean mask to identify rows where the 'Status' contains 'Failed'
# case=False → Case insensitive match
# na=False → Treat NaN as False (avoid errors)

mask_fail = df_final['Status'].str.contains('Failed', case=False, na=False)

# For rows where mask_fail is True, add 1 from 'Designation Level'

df_final.loc[mask_fail, 'Designation Level'] += 1

# 2) Drop rows where the Designation Level is above 4 and the status is Failed.
mask_to_delete = (
    (df_final['Designation Level'] > 4) &
    (df_final['Status'] == 'Failed')
)

# Keep all rows EXCEPT those where mask_to_delete is True
# Reset index (drop=True → removes old index)
df_final = df_final[~mask_to_delete].reset_index(drop=True)

```

```python
'''Task 7 
Add “Mr.” and “Mrs.” to the first name column and drop the gender column. '''


# Update 'First Name' by adding a title based on 'Gender'
# → If Gender is 'M', add prefix 'Mr. '
# → Else (Gender is 'F'), add prefix 'Mrs. '
df_final['First Name'] = np.where(df_final['Gender'] == 'M', 'Mr. ' + df_final['First Name'], 'Mrs. ' + df_final['First Name'])

# Drop the 'Gender' column since it's no longer needed.
df_final.drop(columns= ['Gender'], inplace = True)

```

```python
'''Task 8 
Promote designation level by 1 for the employees whose age is more than 29 years using IF condition. '''


# Update 'Designation Level' by subtracting -1 if 'Age' > 29
# For each row:
# → If Age > 29 → Designation Level +1
# → Else → Keep Designation Level unchanged

df_final['Designation Level'] = (
    df_final.apply(
        lambda row: row['Designation Level'] - 1 if row['Age'] > 29 else row['Designation Level'],
        axis=1 # Apply function row-wise
    )
)
    
```

```python
''' Task 9 
Add the cost of all projects for each Employee and save it in new dataframe “TotalProjCost” with three columns ID, First Name, and Total cost''' 

# Group by 'ID' and 'First Name', then sum the 'Cost' values for each group
TotalProjCost = df_final.groupby(['ID','First Name'])['Cost'].sum().reset_index() 

TotalProjCost.columns = ['ID','First Name', 'Total Cost'] # Rename columns to 'ID', 'First Name', and 'Total Cost'
```
